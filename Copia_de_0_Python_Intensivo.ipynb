##### **4.1 ¿Qué son las librerías?**

- Las librerías de programación hacen referencia funciones que han sido definidas previamente; de manera que al importar una librería estamos trayendo a nuestro entorno de trabajo todas las funciones que contiene esa librería.

- Un ejemplo de función de mean(), de la librería numpy, la cual recibe una lista de números entre los paréntesis, y calcula el promedio de esa lista: mean([5,10,5,20]).

- Así, al ejecutar esta función con la lista dentro, se activa un bloque de código que toma los números que le pasamos a la función y ejecuta un bloque de código creado previamente para devolvernos el promedio, necesidad de sumar los valores de la lista y dividirlos por el número de valores, manualmente.



##### **4.2 Librerías más usadas en el análisis de datos**



- **Pandas**
Se trata de una de las librerías más utilizadas para el tratamiento de datos en Python. Esta herramienta de análisis y manipulación de datos se caracteriza, entre otros aspectos, por definir nuevas funcionalidades de datos basadas en los arrays de la librería NumPy. Permite leer y escribir fácilmente ficheros en formato CSV, Excel y especificar consultas a bases de datos SQL.

Numpy es una librería en la que se define un tipo de dato que representa matrices multidimensionales, equivalentes a las matrices del R. Además incluye algunas funcionalidades básicas para trabajar con ellas. Numpy es una librería con una amplia tradición en Python, es estable y muy rápida.

Esta librería presenta una colección de algoritmos matemáticos y funciones construidas sobre la extensión de NumPy. Incluye módulos de extensión para Python sobre estadística, optimización, integración, álgebra lineal o procesamiento de imágenes, entre otros.

- **Matplotlib**
Matplotlib es una librería de Python especializada en la creación de gráficos en dos dimensiones. Permite crear y personalizar los tipos de gráficos más comunes

- **Seaborn**
Seaborn es otra librería de visualización de datos para Python desarrollada sobre matplotlib. Ofrece una interfaz de alto nivel para la creación de gráficas más atractivas que matplotlib.

#### **5. Creación de un DataFrame a partir de listas**



Volvamos a ver las listas que creamos anteriormente para crear un DataFrame (tabla de datos) a partir de éstas.

```python
# creo una lista con nombres y guardo en variable nombres
nombres = ["tom","luis","juan","pedro"]

# creo una lista con edades guardo en variable edades
edades = [32,23,25,32]

# creo una lista con alturas guardo en variable alturas
alturas = [178,155, 182, 173]
```

```python
# imprimo en pantalla el contenido de la variable nombres
print(nombres)

# imprimo en pantalla el contenido de la variable edades
print(edades)

# imprimo en pantalla el contenido de la variable alturas
print(alturas)
```

```python
zip(nombres,edades,alturas)
```

```python
# importamos la librería pandas como pd porque usaré una función (método) de ésta: DataFrame
import pandas as pd

# definamos un dataframe (tabla de datos) a partir de las listas con el método pd.DataFrame, que recibe información y la convierte en una tabla
df = pd.DataFrame(zip(nombres,edades,alturas),
                  columns=['Nombre','Edad', 'Altura'])
```

```python
df
```

**Fíjense, un dataframe está compuesto por:**
- columnas (variables: Nombre, Edad, Altura)
- filas (registros: ejemplo primera fila = tom, 4, 32)
- un índice (registra cada fila = 0, 1, 2, 3)
- y celdas (contienen los valores = luis o 6 o 22, etc..)

##### **7.3 Cambio de índice del DataFrame**

- Los índices de los dataframes son muy útiles porque nos permiten visualizar el resto de los valores de las disintas columnas del df.
- Es por eso que en ocasiones, querremos cambiar los índices numéricos que se generan por default (0,1,2,3 en nuestro df), por los valores de una columna.
  - por ejemplo, puedo hacer que la columna nombre pase a ser mi índice:

```python
df
```

```python
# el método set index recibe como parámetro el nombre de la columna qdue queremos dejar como índice del dataframe
df.set_index("Nombre") #indice
```

```python
# pero cuidado, algunos métodos, entre ellos set_index, requieren del parámetro inplace=True para que el cambio se guarde en memoria
df
```

```python
# esta vez sí pasamos el parámetro inplace con el argumento True para que el cambio sea permanente
df.set_index ("Nombre", inplace =True)
df
```

Otra ventaja de cambiar tu índice por una columna, es que luego, a la hora de graficar una columna, tu índice dará valores al eje x de tus gráficos.. ya veremos esto en detalle..

##### **7.4 Selección de fila/s por "nombre" del indice con loc**
